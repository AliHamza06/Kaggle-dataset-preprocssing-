# removing Null values

```python
df.isnull().sum()
```

## WE are removing nullvalues in name,hosthame as well also replacing  last_review   10052
### reviews_per_month  10052 to 0 

```python


df = df.dropna(subset=['name', 'host_name'])

# Fill missing values by -0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)
df['last_review'] = df['last_review'].fillna(0)
```

```python
df.isnull().sum()
```

```python
# now dublicates
```

```python
df.drop_duplicates(inplace=True)

df.duplicated().sum()
```

```python
df['last_review'] = pd.to_datetime(df['last_review'], errors='coerce')
```

```python
# data types 
```

```python
df.dtypes
```

# statisticaal info

```python
df['price'].describe()
```

```python
# logical errors
df = df[df['price'] > 0]
df = df[df['minimum_nights'] > 0]
```

# str.strip() removes any extra spaces from the start and end of text in each cell.

```python
df['neighbourhood_group'] = df['neighbourhood_group'].str.strip()
df['room_type'] = df['room_type'].str.strip()
```

```python
df['neighbourhood_group'].value_counts()
```

### now resewting index as we have removed sum nll instances

```python
df.reset_index(drop=True, inplace=True)
df
```
